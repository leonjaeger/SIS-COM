```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
from scipy.io.wavfile import write
```

```python
mps=500
fec_portadora=100
fec_moduladora=1
duracion=3
k=15.0
```

```python
muestra=fec_moduladora/mps
t=np.arange(0,1,muestra)
```

```python
moduladora=np.sin(2*np.pi*fec_moduladora*t)
```

```python
modulada=np.sin(2*np.pi*(fec_portadora+k*moduladora)*t)
```

```python
plt.plot(t,modulada)
```
